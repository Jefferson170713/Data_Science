# Concat.
    Serve para concatenar elementos.

```python
import pandas as pd
```

```python
df_1 = pd.DataFrame(
    {
        'A': ['A0','A1','A2','A3'],
        'B': ['B0','B1','B2','B3'],
        'C': ['C0','C1','C2','C3'],
        'A': ['D0','D1','D2','D3']
    }, index= [0,1,2,3]
)
```

```python
df_2 = pd.DataFrame(
    {
        'A': ['A0','A1','A2','A3'],
        'B': ['B0','B1','B2','B3'],
        'C': ['C0','C1','C2','C3'],
        'A': ['D0','D1','D2','D3']
    }, index= [4,5,6,7]
)
```

```python
df_3 = pd.DataFrame(
    {
        'A': ['A0','A1','A2','A3'],
        'B': ['B0','B1','B2','B3'],
        'C': ['C0','C1','C2','C3'],
        'A': ['D0','D1','D2','D3']
    }, index= [8,9,10,11]
)
```

#### Veja que vai juntas os Data frames em Df.

```python
df = pd.concat([df_1, df_2, df_3])
df
```

```python
df = pd.concat([df_1, df_2, df_3], axis=1)
df
```
