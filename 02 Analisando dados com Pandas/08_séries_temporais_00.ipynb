## Séries temporáris no pandas.
    No pandas podemos pegar, trabalhar e importar dados em tempo real, para isso que serve série temporal.

```python
import pandas as pd
```

```python
datas = pd.date_range(start='01/10/2022', periods=150)
datas
```

### Detalhe:
    Observar o formato da variável datas = DatetimeIndex.
    Agora vamos passsar para um dataFrame.

```python
df = pd.DataFrame(datas)
df
```

```python
df.info()
```

```python
df.infer_objects()
```

### Agora vamos adcionar mais um coluna.

```python
num = range(1,151)
df['Número'] = num
df
```

```python
df_1 = pd.DataFrame(num, columns=['Número'], index=datas)
df_1
```

```python
datas1 = pd.date_range(start='09/06/2022', periods=60)
num2 = range(151,211)
df_2 = pd.DataFrame(num2, columns=['Número'], index=datas1)
df_3 = pd.concat([df_1, df_2])
df_3
```
