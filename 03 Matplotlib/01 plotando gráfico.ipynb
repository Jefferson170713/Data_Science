```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('GDP.csv')
df
```

```python
df['GDP'] = df[' GDP_pp '].apply(lambda x : x.replace(',',''))
df['GDP'] = df['GDP'].apply(lambda x : float(x))
type(df['GDP'][1])
print(df['GDP'])
```

```python
df['YEAR'] = df['Year'].apply(lambda x : x.split('/')[2])
df['YEAR'] = df['YEAR'].apply(lambda x : int(x))
df['YEAR']
```

```python
x = df[df['Country'] == 'Brazil']['GDP']
y = df[df['Country'] == 'Brazil']['YEAR']
plt.plot( df[df['Country'] == 'Brazil']['YEAR'], df[df['Country'] == 'Brazil']['GDP'], 'g--')
plt.xlabel('YEAR')
plt.ylabel('GDP')
#plt.title('PIB do Brazil')

plt.plot( df[df['Country'] == 'United States']['YEAR'], df[df['Country'] == 'United States']['GDP'], 'b--')
plt.xlabel('YEAR')
plt.ylabel('GDP')
plt.title('PIB (EUA x Brazil)')
```

```python
plt.plot( df[df['Country'] == 'United States']['YEAR'], df[df['Country'] == 'United States']['GDP'], 'b--')
plt.xlabel('YEAR')
plt.ylabel('GDP')

plt.plot( df[df['Country'] == 'China']['YEAR'], df[df['Country'] == 'China']['GDP'], 'r--')
plt.xlabel('YEAR')
plt.ylabel('GDP')
plt.title('PIB (EUA x China)')
```
