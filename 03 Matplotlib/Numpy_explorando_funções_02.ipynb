```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(25)
y = np.random.random(25)
```

```python
plt.plot(x, y)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, 'r--')

plt.subplot(1, 2, 2)
plt.plot(y, 'b--')
```

```python
plt.subplot(1, 2, 1)
plt.hist(x)

plt.subplot(1, 2, 2)
plt.hist(y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # esquerda, inferior, largura e altura.4
axes.plot(x, 'r')
axes.plot(y, 'b')
axes.set_xlabel('eixo x')
axes.set_ylabel('eixo y')
axes.set_title('TÍTULO')
```
