```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(1, 5, num=10)
a
```

### A função.
    - Cria valores igualmente distribuido entre um determinado intervalo de valores.

```python
plt.plot(a)
```

```python
b = -a ** 2
b
```

```python
plt.plot(b)
```

```python
c = a ** 2
plt.plot(c)
```

```python
plt.plot(c, 'bo')
plt.plot(b, 'go')
plt.plot(a, 'ro')
```

```python
plt.plot(c, a)
```

#### Agora com subplots

```python
plt.subplot(1, 3, 1)
plt.plot(a, 'r--')
plt.subplot(1, 3, 2)
plt.plot(c, 'g--')
plt.subplot(1, 3, 3)
plt.plot(b)
```
