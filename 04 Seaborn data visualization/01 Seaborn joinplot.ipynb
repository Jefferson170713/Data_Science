```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['total_real'] = df['total_bill'] * df['size'] - (df['tip'] * df['size'])
df['total_sex'] = df['total_bill'] * df['size']
df
```

```python

x1 = (df[df['sex'] == 'Male']['total_bill'] * df[df['sex'] == 'Male']['size'] - (df['tip'] * df['size']))
x2 = (df[df['sex'] == 'Female']['total_bill'] * df[df['sex'] == 'Female']['size'] - (df['tip'] * df['size']))

x1
```
