```python
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("C:/Users/55859/Downloads/Cities_of_Brazil.csv")
df.head(2)
```

```python
fig = px.scatter_mapbox(df, lat="lat_municipio", lon="long_municipio", hover_name="nm_municipio", hover_data=["nm_municipio", "nm_uf"],
                        color_discrete_sequence=['darkolivegreen'], zoom=3, height=300)
```

```python
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
