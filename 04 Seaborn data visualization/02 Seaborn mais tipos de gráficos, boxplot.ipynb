```python
import seaborn  as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head(3)
```

#### Repare que passei os dados.
    - O ponnto de preto, está sendo calculado pela média.
    - Poderia passar também o desvio padrão. como no próximo gráfico.

```python
sns.barplot(x= 'sex', y = 'total_bill', data=df)
```

```python
sns.barplot(x= 'sex', y = 'total_bill', data=df, estimator=np.std) # por desvio padrão.
```

### Mostrando em gráficos pela diferença entre variáveis categoricas.
    - Exemplo disso seria a variável sex. Vai devolver em forma de gráfico e por contagem.

```python
sns.countplot(x='smoker', data=df, hue='sex') # fumante e não fumante
```

#### Boxplot.
    vai mostrar como os valores e média estão distribuidos neste conjunto de dados.

```python
df.head(2)
```

##### Gráfico 1

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow', hue='sex')
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow') # separando por sexo.
```

#### Gráfico 2 com inversão de eixos.

```python
sns.boxplot(x='total_bill', y='day', data=df, palette='rainbow')
```

```python
sns.boxplot(x='total_bill', y='day', data=df, hue='sex', palette='rainbow')
```

#### Violinplot.

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', palette='rainbow', split=True)
```

#### Swarplot.
    Mostra a distribuição normal dos dados. Sem média ou quartis.

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, hue='sex')
```
