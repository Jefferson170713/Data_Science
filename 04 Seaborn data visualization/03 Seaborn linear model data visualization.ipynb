```python
df = sns.load_dataset('tips')
df.head(3)
```

```python
df.info()
```

```python
df.columns
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

### Com diviso por sexo.

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex')
```

### Com marcadores.

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', markers=['o','x'])
```

### Com almento na horizintal.
    com o parmentro aspect.

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', markers=['o','x'], aspect=3)
```

### Com o tamanho, nome antigo size, agora  height.

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', markers=['o','x'], aspect=3, size=5)
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', markers=['o','x'], aspect=3, height=4.5)
```
