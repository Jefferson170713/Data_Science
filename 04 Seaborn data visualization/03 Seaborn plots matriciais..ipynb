```python
import seaborn as sns
```

```python
dado = sns.load_dataset('flights')
dado.head(3)
```

```python
voos = dado.pivot_table(index='year', columns='month', values='passengers',sort=False)
voos
```

```python
voos = dado.pivot_table(index='year', columns='month', values='passengers',sort=True)
voos
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
t = tips.pivot_table(index='sex', columns='day', values='total_bill',sort=True)
t.info()
```

### Agora vamos aos gráficos matriciais.
#### Mas para isso, devemos primeiro converter nossos dados em uma matriz com a função pivot_tabble.
#### Após configurarmos nosso índice e nossas colunas, é o momento de usar uma das funções principais do SEABORN.
# sns.heatmap().
    A fução plota um gráfico de três eixos. x, y e z=('sendo igual a cor que irá mostrar').

```python
sns.heatmap(t)
```

```python
sns.heatmap(voos)
```

##### Olhando os parâmetros.

```python
sns.heatmap(voos, robust=False,)
```

### Almentando o tamanho do gráfico com o matplotlib.

```python
import matplotlib.pyplot as plt
```

```python
fig , eixo = plt.subplots(figsize=(20,7))
sns.heatmap(voos, ax=eixo)
```

```python
figg, ex = plt.subplots(figsize=(15, 5))
sns.heatmap(t, ax=ex)
```

#### Agora vamos colocar valores nos gráficos.
    Função 'annot', que pode ser (False ou True), mas vamos alterar. Essa função mostra os valores.
    Função fmt diz repeito as casas decimais, se mostra números inteiro ou não. fmt='.0f'.

```python
fig , eixo = plt.subplots(figsize=(20,7))
sns.heatmap(voos, ax=eixo, annot=True, fmt='.1f')
```

```python
figg, ex = plt.subplots(figsize=(15, 5))
sns.heatmap(t, ax=ex, annot=True, fmt='0.2f')
```

```python
fig , eixo = plt.subplots(figsize=(20,7))
sns.heatmap(voos, ax=eixo, annot=True, fmt='.1f',linewidths=0.5, cmap='nipy_spectral')
```
