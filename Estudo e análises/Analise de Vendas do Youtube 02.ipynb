### 2.1 - Tratamento das variáveis Categoricas.

```python
# 2.1.0 Tratatando da coluna Localidade e mudando para duas novas colunas.
# 2.1.2 - País
pais = categoricas['Localidade'].apply(lambda x : x.split(' - ')[0])
```

### 3.0 - Tratamento das variáveis Numéricas.
    Relção de preço, custo e lucro.
    3.1 - Preço unitário foi no quanto foi vendido para o consumidor final.
    3.2 - Custo unitário foi quanto essa mercadoria custou para o dono do negócio.
    3.3 - Valor real -> Quantas unidades reais foram vendidas
    3.3 - O custo real por unidade vendida.
    3.4 - Novas colunas -> 'Venda Real', 'Custo real'.
    3.5 - Nova varial -> 'lucro'
######    3.6 - OBS: lembrando que as variáveis 'Venda Real', 'Custo real' não entraram para o Modelo de Machine learning lá na frente.

```python
numericas.head()
```

```python
# 3.1 - Preço unitário foi no quanto foi vendido para o consumidor final.
numericas['Valor das Vendas'] = numericas['PrecoUnitario'] * numericas['Qtd. Vendida']

# 3.2 - Custo unitário foi quanto essa mercadoria custou para o dono do negócio.
numericas['Custo Real'] = numericas['Custo Unitário'] * numericas['Qtd. Vendida']

# 3.5 - Nova varial -> 'lucro'
numericas['Lucro'] = numericas['Valor das Vendas'] - numericas['Custo Real']

numericas.head()
```

###### 4.1 - Produto

```python
#categoricas['Produto'].unique()
for num, cate in enumerate(categoricas['Produto'].unique()):
    print(f'{num+1} - {cate}')
```

```python
PRODUTO = pd.get_dummies(categoricas['Produto'])
PRODUTO
```

###### 4.2 - Categoria

```python
for num, cate in enumerate(categoricas['Categoria'].unique()):
    print(f'{num+1} - {cate}')
```

```python
CATEGORIA = pd.get_dummies(categoricas['Categoria'])
CATEGORIA
```

###### 4.3 - Marca

```python
for num, ma in enumerate(categoricas['Marca'].unique()):
    print(f'{num+1} - {ma}')
```
