```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_sp = pd.read_csv('sao-paulo-properties-april-2019.csv')
```

```python
df_sp.info()
```

```python
df_rent = df_sp[df_sp['Negotiation Type'] == 'rent']
df_rent.head(2)
```

```python
df_cleaned = df_rent.drop(['New', 'Negotiation Type','Property Type'], axis=1)
df_cleaned.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_enconder = OneHotEncoder()
```

```python
housing_cat_1hot = cat_enconder.fit_transform(df_cleaned[['District']])
```

```python
housing_cat_1hot
```

```python
housing_cat_1hot.toarray()
```

```python
one_hot = pd.get_dummies(df_cleaned['District'])
one_hot
```

```python
df = df_cleaned.drop('District', axis=1)
df.head(2)
```

```python
df = df.join(one_hot)
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y = df['Price'] 
#Y.sort_values(ascending=True)
```

```python
X = df.loc[:, df.columns != 'Price']
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3) # Treino e, 70% e teste em 30% do nosso df.
```

```python
x_train
```

```python
y_train
```

```python
# LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train) # Aqui vai usar a função custo para otimização do modelo
```

```python
alguns_dados = x_train.iloc[:10]
algumas_label = y_train.iloc[: 10]
```

```python
#print('Predições: ', lin_reg.predict(alguns_dados))
print(f'*'*25, ' Predições', '*'*25)
for a, b in enumerate(lin_reg.predict(alguns_dados)):
    print(a + 1,' : ', f'{b:.02f}')
```

```python
#print('Labels', algumas_label.values)
print(f'*'*25, ' Labels ', '*'*25)
for a, b in enumerate(algumas_label.values):
    print(a + 1,' : ', f'{b:.02f}')
```

```python
lin_reg.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predts = lin_reg.predict(x_train)

lin_mse = mean_squared_error(y_train, predts)

lin_mse = np.sqrt(lin_mse)

lin_mse
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(lin_reg, x_train, y_train, scoring='neg_mean_squared_error', cv=10)
```

```python
lin_rmse_scores = np.sqrt(-score)
```

```python
def diskplay_score(scores):
    for a, b in enumerate(scores):
        print(f'Score {a + 1} : ', f'{b:.05f}')
    #print('Score :', scores)
    print()
    print('Média : ', scores.mean())
    print('Desvio padrão : ', scores.std())
```

```python
diskplay_score(lin_rmse_scores)
```
