### Agora vamos importar o nosso modelo classificador.
    Depois vamos instânciar, pois é uma classe.

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_class = SGDClassifier()
```

### Agora passamos as nossas fits.
    Nosso x de treino e o nosso booleanos, para o modelo classificar quais deles são verdadeiros ou quais são falsos.
    Lembrando que escolhemos o número 7

```python
sgd_class.fit(x_train, y_train_7)
```

### Obs:
    print('Classe predita pelo modelo: ', sgd_class.predict([x_train[n]]))
    Na etapa de predição, ele só vai aceitar e passarmos um dataframe.

```python
n = 5
plt.imshow(x_train[n].reshape(28, 28), cmap='binary_r')
print('Classe real: ',y_train_7[n])
print('Classe predita pelo modelo: ', sgd_class.predict([x_train[n]]))
plt.show()
```
