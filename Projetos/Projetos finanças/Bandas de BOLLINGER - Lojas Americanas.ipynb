```python
bitcoin = yf.download('BTC-USD', start='2021-01-01')
```

```python
bitcoin
```

```python
bitcoin['Close'].plot(figsize= (20,8))
```

```python
bitcoin['MMS_20'] = bitcoin['Close'].rolling(window = 20).mean()

bitcoin['Desvio_padrao'] = bitcoin['Close'].rolling(window = 20).std()
```

```python
bitcoin
```

```python
bitcoin['Banda_Superior'] = bitcoin['MMS_20'] + (bitcoin['Desvio_padrao'] * 2)
bitcoin['Banda_Inferior'] = bitcoin['MMS_20'] - (bitcoin['Desvio_padrao'] * 2)
bitcoin
```

```python
bitcoin[['Close', 'MMS_20', 'Banda_Superior', 'Banda_Inferior']].plot(figsize = (20, 8))
plt.title('Bandas de BOLLINGER', fontsize = 25)
```

```python
bitcoin['Sinal_compra'] = np.where(bitcoin['Banda_Inferior'] > bitcoin['Close'], bitcoin['Close'], np.nan)
bitcoin['Sinal_venda'] = np.where(bitcoin['Banda_Superior'] < bitcoin['Close'], bitcoin['Close'], np.nan)
bitcoin
```

```python
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(1,1,1)

ax.plot(bitcoin.index, bitcoin['Close'], linewidth = 2, label = 'Bitcoin')
ax.plot(bitcoin.index, bitcoin['MMS_20'], linewidth = 1, label = 'Média móvel simples - 20 dias')
ax.plot(bitcoin.index, bitcoin['Banda_Superior'], linewidth = 1, label = 'Banda Superior')
ax.plot(bitcoin.index, bitcoin['Banda_Inferior'], linewidth = 1, label = 'Banda Superior')

ax.scatter(bitcoin.index, bitcoin['Sinal_compra'], marker = '^', linewidth = 3, color = 'green', label = 'Sinal de Compra')
ax.scatter(bitcoin.index, bitcoin['Sinal_venda'], marker = 'v', linewidth = 3, color = 'red', label = 'Sinanl de Venda')

plt.title('Ações do BITCOIN com Bandas de BOLLINGER', fontsize = 25)
plt.legend()
plt.show()
```
